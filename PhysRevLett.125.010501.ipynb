```python
import numpy as np
from numpy import log, log10, sin, cos, exp, sqrt, pi
import pandas as pd
import glob as glob

# plotting
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from matplotlib import cm

import json

# warnings
import warnings
warnings.simplefilter('ignore')

def arraycenter(x):
    """
        returns centered array for histograms
    """
    return np.array([(x[i]+x[i+1])/2 for i in range(len(x)-1)])
```

```python
# fluctuations in the data will be due to uncertainty in selecting points in WPD
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(6,4))

### fig 3
f = open('dataWPD/Fig3/wpd.json')
data = json.load(f)
# 
datatmp = data['datasetColl'][0]['data']
datatmp2 = np.array([datatmp[i]['value'] for i in range(len(datatmp))])
indxs = np.argsort(datatmp2[:,0])
plt.plot(datatmp2[:,0][indxs],datatmp2[:,1][indxs],'.', color='lightblue', markersize=2, label='Ideal WPD')
# 
datatmp = data['datasetColl'][1]['data']
datatmp2 = np.array([datatmp[i]['value'] for i in range(len(datatmp))])
indxs = np.argsort(datatmp2[:,0])
plt.plot(datatmp2[:,0][indxs],datatmp2[:,1][indxs],'.', color='gold', markersize=2, label='Dissipative WPD')
# 
#plt.xlim([0,15])
plt.title(r'Fig 3')
plt.xlabel(r'Time')
plt.ylabel(r'Correlation')
plt.legend()
plt.grid()
plt.tight_layout()
#plt.savefig('fig3_dataWPD.pdf', transparent=True)
plt.show()
```
